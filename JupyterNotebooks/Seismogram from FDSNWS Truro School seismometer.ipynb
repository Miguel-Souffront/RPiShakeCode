```python
%matplotlib inline
import obspy
import matplotlib as plt
from obspy.clients.fdsn import Client
client = Client(base_url='https://fdsnws.raspberryshakedata.com/')
from obspy import UTCDateTime
t = UTCDateTime("2020-01-05T04:32:53.000") # 2020-01-05 04:32:53 (UTC)64.701N 17.752W 8.2km depth
[ sbef, saft ] = [ -120, 1500 ]
st = client.get_waveforms('AM', 'R7FA5', '00', 'EHZ',t - sbef, t + saft, attach_response = True) # Truro School
st.merge(method=0, fill_value='latest')
st.detrend(type='demean')
st.filter("bandpass", freqmin=1, freqmax=2.0, corners=2, zerophase=False)
st.plot(linewidth=0.75)
```
