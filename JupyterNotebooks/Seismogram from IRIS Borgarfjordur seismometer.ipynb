```python
%matplotlib inline
import obspy
import matplotlib as plt
from obspy.clients.fdsn import Client
client = Client("IRIS")
from obspy import UTCDateTime
t = UTCDateTime("2020-01-05T04:32:53.000") # 2020-01-05 04:32:53 (UTC)64.701N 17.752W 8.2km depth
[ sbef, saft ] = [ 70, 210 ]
st = client.get_waveforms( "II", "BORG", "00", "BHZ", t - sbef, t + saft, attach_response = True)
st.filter("bandpass", freqmin=0.6, freqmax=5.0, corners=2, zerophase=False)
st.plot(linewidth=0.75)
```
